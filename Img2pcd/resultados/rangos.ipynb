```python
import open3d
import numpy as np
```

```python
pcd_nuevo = open3d.io.read_point_cloud("./IMG_2632_nuevo.ply")
pcd_viejo = open3d.io.read_point_cloud("./IMG_2632_viejo.ply")
```

```python
z_nuevo = np.asarray(pcd_nuevo.points)[:, 0]
z_viejo = np.asarray(pcd_viejo.points)[:, 0]
#graph boxplot
import matplotlib.pyplot as plt
plt.boxplot([z_viejo, z_nuevo])
plt.xticks([1, 2], ['viejo (vinvino02/glpn-nyu)', 'nuevo(Depth-Anything-V2-Small-hf)'])
plt.ylabel('Valor X')
plt.xlabel('Modelo')
plt.show()
```
