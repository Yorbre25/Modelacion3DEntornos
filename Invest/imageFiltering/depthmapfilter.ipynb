```python
import cv2
import matplotlib.pyplot as plt

# Leer la imagen
dm = cv2.imread("./../Data/depth/bottle_center_depth.png")
dm = cv2.cvtColor(dm, cv2.COLOR_BGR2RGB)  # Convertir de BGR a RGB

# Aplicar el filtro bilateral
# cv2.bilateralFilter(src, d, sigmaColor, sigmaSpace)
# d: Dimetro de cada pxel en el vecindario.
# sigmaColor: Desviacin estndar en el espacio de color.
# sigmaSpace: Desviacin estndar en el espacio de coordenadas.
filtered_dm = cv2.bilateralFilter(dm, d=9, sigmaColor=75, sigmaSpace=75)

# Mostrar la imagen original y la filtrada
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title('Original')
plt.imshow(dm)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Bilateral Filter')
plt.imshow(filtered_dm)
plt.axis('off')

plt.show()
```
